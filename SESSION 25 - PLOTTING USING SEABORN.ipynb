### 1. Relational Plot

- To see the statistical relation between 2 or more variables.
- Bivariate Analysis
- Plots under this section
    - scatterplot
    - lineplot
    - relplot

```python
import seaborn as sns; sns.set_theme()
import pandas as pd
import plotly.express as px # for toy dataset
```

```python
tips = sns.load_dataset('tips')
tips
```

**scatter plot**
- Draw a scatter plot for numerical data
- **Syntax: sns.scatterplot(data, x, y, hue, size, style, markers=True,
  legend='auto', col, row, col_wrap)**

```python
# scatter plot > axes level function
sns.scatterplot(data=tips, x='total_bill', y='tip', 
                hue='sex', style='time', size='size')
```

**replot:**
- Figure-level interface for drawing relational plots onto a FacetGrid (a multi-axes grid with subplots visualizing the distribution ).
  
- This function provides access to several different axes-level functions that show the relationship between two variables.

- **Syntax: sns.relplot(data, x, y, kind='scatter', hue, size, style, markers=True,
  legend='auto', col, row, col_wrap)**

- The kind parameter selects the underlying axes-level function to use:
    - scatterplot() (with kind="scatter"; the default)
    - lineplot() (with kind="line")

```python
# relplot > Figure level function  > square shape
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', 
            hue='sex', style='time', size='size')
```

**line plot**
- If you wnat continuty in graph then use line plot
- Most of use in time series data like time, years, months, day, hours etc.
- **Syntax: sns.lineplot(data, x, y, hue, size, style, markers=True, legend='auto')**

```python
gap = px.data.gapminder()
gap
```

```python
# life Exp of India
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# life Exp of India over the year
# Axes level function
sns.lineplot(temp_df, x='year', y='lifeExp')
```

```python
# using replot
sns.relplot(temp_df, x='year', y='lifeExp', kind='line')
```

```python
# hue & style
temp_df = gap[gap['country'].isin(['India','Brazil','Japan','Germany'])]
```

```python
# hue & style with relplot
sns.relplot(temp_df, x='year', y='lifeExp',kind='line', hue='country')
```

```python
# hue & style with lineplot
sns.lineplot(temp_df, x='year', y='lifeExp',hue='country', 
             style='continent', size='continent')
```

**facet plot**
- In fact plots draw the same type of plot
- Work on figure level function
- Work with relplot
- Does Not Work with Scatterplot and Lineplot

```python
#sns.relplot(tips, x='total_bill', y='tip', kind='scatter', col='sex', row='day')
sns.relplot(tips, x='total_bill', y='tip', kind='line', col='sex', row='day')
```

```python
# col_wrap parameter
# lifeExp vs GDP 
sns.relplot(gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=3)
```

### 2. Distribution Plots
- A distribution plot is a way to visualize the distribution of a univariate set of observations.
- It provides a convenient way to understand the underlying distribution of a variable and to identify patterns such as skewness, central tendency, and spread.
- Used for univariate analysis (focus on only one column)
- used to find out the distribution
- Range of the observation
- Central Tendency - Mean , median, mode
- is the data bimodal? - Picks in data
- Are there outliers? - Find outliers

- **Plots under distribution plot**
    - displot
    - histplot
    - kdeplot
    - rugplot
 
- **figure level :** displot
- **axes level:** histplot, kdeplot, rugplot

**displot**
- Figure-level interface for drawing distribution plots onto a FacetGrid.
- **Syntax: sns.displot(data, x, y, hue, row, col, kind='hist', rug=False, bins, legend=True, color, col_wrap)**

```python
sns.displot(tips, x='total_bill', kind='hist')
```

**hisplot**
- Plot univariate or bivariate histograms to show distributions of datasets.
- **Syntax: sns.histplot(data, x, y, hue, bins='auto', element='bars', fill=True, kde=False, color, legend=True)**

```python
# Plotting univariate histograms
sns.histplot(data=tips, x='total_bill', bins=10)
```

```python
# Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram. 
# Discrete bins are automatically set for categorical variables

# countplot
sns.displot(tips, x='day', kind='hist')
```

```python
# hue and element parameters
sns.displot(tips, x='tip', kind='hist', hue='sex', element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x='age', kind='hist', element='step', hue='sex')
```

```python
# facet plot using col and row -> not work on histplot function
sns.displot(data=tips, x='tip', kind='hist', col='sex', element='step', hue='sex')
```

**Bivariate histogram**
- A bivariate histogram bins the data within rectangles that tile the plot and then shows the count of observations within each rectangle with the fill color

```python
# Bivariate histogram with histplot and displot
sns.histplot(tips, x='total_bill', y='tip')
#sns.displot(tips, x='total_bill', y='tip', kind='hist')
```

**kdeplot**
- Plot univariate or bivariate distributions using kernel density estimation (kde).
- Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
- It will give you the density of the probability in data.
- **Syntax: sns.kdeplot(data, x, y, hue, weights, color, fill=False, legend=True)**

```python
# kdeplot
sns.kdeplot(tips, x='total_bill', hue='sex', fill=True)
```

**Rugplot**
- Plot marginal distributions by drawing ticks along the x and y axes.
- This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
- Generally use of this plot with others plots 
- **Syntax: sns.rugplot(data, x, y, hue, legend=True)**

```python
sns.kdeplot(tips, x='total_bill')
sns.rugplot(tips, x='total_bill', hue='sex')
```

**Bivariate Kdeplot**
- A bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian

```python
# Bivariate kde with kdeplot and displot
sns.kdeplot(tips, x='total_bill', y='tip', fill=True)
#sns.displot(tips, x='total_bill', y='tip', kind='kde')
```
